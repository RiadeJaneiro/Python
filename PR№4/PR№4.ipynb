```python
# РПОС_ПР№4.2.1
# Выполнил: Городнянская М.С.
# Группа: МС-32

import numpy as np
import math
def sgn(x):
    """Вернуть результат сигнум-функции.

    Параметры:
        - x (float): аргумент x.

    Результат:
        float: значение функции.
    """
    return -1 if x < 0 else 1 if x > 0 else 0

x = float(input('Введите x: '))
y = float(input('Введите y: '))

z = round(float((sgn(x) + y**2) / (sgn(y) - math.sqrt(abs(x)))), 2)

print("Ответ:", z)

# --------------
# Пример вывода:
#
# Введите x: -9
# Введите y: 0
# Ответ: 0.33
```

```python
# РПОС_ПР№4.2.2
# Выполнил: Городнянская М.С.
# Группа: МС-32



def avg(data):
    """Вернуть среднее значение 'data'.

    Параметры:
        !!!.

    Результат:
        !!!.
    """
    return float(sum(data) / len(data))


def cleared_data(data):
    """Вернуть список без значений None.

    Параметры:
        !!!.

    Результат:
        !!!.
    """
    return [x for x in data if x is not None]


n = int(input('Введите кол-во измерений: '))

# Если с клавиатуры вводится прочерк, измерения нет
# (необходимо добавить в список None)
data = []

for num in range(n):
  num = input()
  if num == '-':
    data.append(None)
  else:
    data.append(float(num))

# Получить очищенный список и среднее значение
clearedData = cleared_data(data)

print('Кол-во измерений:', n)

for i, num in enumerate(data, start = 1):
  print(f'Измерение {i}-е:', int(num) if num is not None else '-')

print('Средняя температура: {:.2f}'.format(avg(clearedData)))

# --------------
# Пример вывода:
#
# Кол-во измерений: 3
# Измерение 1-е: 10
# Измерение 2-е: -
# Измерение 3-е: 20
# Средняя температура: 15.00
```

```python
# РПОС_ПР№4.2.3
# Выполнил: Городнянская М.С.
# Группа: МС-32



def is_lucky(num):
    """Является ли 'num' номером счастливого билета?

    Параметры:
        !!!.

    Результат:
        bool: True - да, False - нет.
    """
    num_even = 0
    num_odd = 0

    for digit in str(num):
      if int(digit) % 2 == 0:
        num_even += 1
      else:
        num_odd += 1

    return num_even == num_odd


def lucky_numbers(a, b):
    """Вернуть список счастливых номеров от 'a' до 'b'.

    Параметры:
        !!!.

    Результат:
        !!!.
    """
    return [num for num in range(a, b + 1) if is_lucky(num)]


a = int(input('Первый номер билета: '))
b = int(input('Последний номер билета: '))

# Вывод должен быть осуществлен в строку с одним пробелом-разделителем
print(*lucky_numbers(a, b))

# --------------
# Пример вывода:
#
# Первый номер билета: 10
# Последний номер билета: 25
# 10 12 14 16 18 21 23 25
```

```python
# РПОС_ПР№4.2.4
# Выполнил: Городнянская М.С.
# Группа: МС-32



def is_leap(year):
    """Является ли 'year' високосным годом?

    Параметры:
        .

    Результат:
        bool: True - да, False - нет.
    """
    if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0):
      return True
    return False


def days(month, year):
    """Вернуть количество дней в месяце 'month' года 'year'.

    Параметры:
        !!!.

    Результат:
        !!!.
    """
    days31 = [1, 3, 5, 7, 8, 10, 12]
    days30 = [4, 6, 9, 11]

    if month in days31:
      return 31
    if month in days30:
      return 30
    elif month == 2:
      if is_leap(year):
        return 29
      else:
        return 28


def previous_date(day, month, year):
    """Вернуть день, месяц, год предыдущего дня.

    Параметры:
        !!!.

    Результат:
        !!!.
    """
    if day == 1:
      if month == 1:
        day = 31
        month = 12
        year -= 1
      else:
        month -= 1
        day = days(month, year)
    else:
      day -= 1

    return day, month, year


def next_date(day, month, year):
    """Вернуть день, месяц, год следующего дня.

    Параметры:
        !!!.

    Результат:
        !!!.
    """
    if day == days(month, year):
      if month == 12:
        day = 1
        month = 1
        year += 1
      else:
        day = 1
        month += 1
    else:
      day += 1

    return day, month, year


day, month, year = map(int, input('День, месяц, год через пробел: ').split())

prevDay, prevMonth, prevYear = previous_date(day, month, year)
nextDay, nextMonth, nextYear = next_date(day, month, year)

print(f'Предыдущий день: {prevDay:02d}/{prevMonth:02d}/{prevYear}')
print(f'Следующий день: {nextDay:02d}/{nextMonth:02d}/{nextYear}')

# --------------
# Пример вывода:
#
# День, месяц, год через пробел: 1 3 2000
# Предыдущий день: 29/02/2000
# Следующий день: 02/03/2000
```

```python
# РПОС_ПР№4.2.5
# Выполнил: Городнянская М.С.
# Группа: МС-32



def is_leap(year):
    """Является ли 'year' високосным годом?

    Параметры:
        .

    Результат:
        bool: True - да, False - нет.
    """
    if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0):
      return True
    return False


def days(month, year):
    """Вернуть количество дней в месяце 'month' года 'year'.

    Параметры:
        !!!.

    Результат:
        !!!.
    """
    days31 = [1, 3, 5, 7, 8, 10, 12]
    days30 = [4, 6, 9, 11]

    if month in days31:
      return 31
    if month in days30:
      return 30
    elif month == 2:
      if is_leap(year):
        return 29
      else:
        return 28


def another_date(day, month, year, delta=1):
    """Вернуть день, месяц, год, отличающийся на 'delta' дней.

    Параметры:
        !!!.

    Результат:
        !!!.
    """

    def previous_date(day, month, year):
        """Вернуть день, месяц, год предыдущего дня.

        Параметры:
            !!!.

        Результат:
            !!!.
        """
        if day == 1:
          if month == 1:
            day = 31
            month = 12
            year -= 1
          else:
            month -= 1
            day = days(month, year)
        else:
          day -= 1

        return day, month, year


    def next_date(day, month, year):
        """Вернуть день, месяц, год следующего дня.

        Параметры:
            !!!.

        Результат:
            !!!.
        """
        if day == days(month, year):
          if month == 12:
            day = 1
            month = 1
            year += 1
          else:
            day = 1
            month += 1
        else:
          day += 1

        return day, month, year


    for _ in range(abs(delta)):
      if delta >= 0:
        day, month, year = next_date(day, month, year)
      else:
        day, month, year = previous_date(day, month, year)

    return day, month, year

day, month, year = map(int, input('День, месяц, год через пробел: ').split())
delta = int(input('Свдиг (может быть отрицательным): '))

anotherDay, anotherMonth, anotherYear = another_date(day, month, year, delta)

print(f'Новый день: {anotherDay:02d}/{anotherMonth:02d}/{anotherYear}')

# --------------
# Пример вывода:
#
# День, месяц, год через пробел: 1 1 2000
# Свдиг (может быть отрицательным): -2
# Новый день: 30/12/1999
#
# День, месяц, год через пробел: 1 1 2000
# Свдиг (может быть отрицательным): 2
# Новый день: 03/01/2000
```

```python
# РПОС_ПР№4.2.6
# Выполнил: Городнянская М.С.
# Группа: МС-32

from functools import reduce

def gcd(first, second):
    """Вернуть НОД целых чисел 'first' и 'second'."""
    while first != 0 and second != 0:
      if first > second:
        first = first % second
      else:
        second = second % first

    return first + second


def lcm(first, second):
    """Вернуть НОК целых чисел 'first' и 'second'."""
    return (first * second) // gcd(first, second)


def gcd_nums(nums):
    """Вернуть НОД чисел из списка 'nums'.

    Параметры:
        !!!.

    Результат:
        !!!.
    """
    return reduce(gcd, nums)


def lcm_nums(nums):
    """Вернуть НОК чисел из списка 'nums'.

    Параметры:
        !!!.

    Результат:
        !!!.
    """
    return reduce(lcm, nums)


nums = list(map(int, input('Введите числа через пробел: ').split()))

print('НОД =', gcd_nums(nums))
print('НОК =', lcm_nums(nums))

# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```

```python
# РПОС_ПР№4.2.9
# Выполнил: Городнянская М.С.
# Группа: МС-32



LETTERS_EX = {10: "A", 11: "B", 12: "C", 13: "D", 14: "E", 15: "F"}
DIGITS_EX = {"A": 10, "B": 11, "C": 12, "D": 13, "E": 14, "F": 15}


def to_base(number, base):
    """Преобразовать десятичное число 'number' в с.с. 'base'.

    Параметры:
        number (int): десятичное число;
        base (int): система счисления.

    Результат:
        str: число в с.с. 'base'."""
    if number == 0:
      return '0'

    result = ''
    while number > 0:
      remainder = number % base
      if remainder > 9:
        result = LETTERS_EX[remainder] + result
      else:
        result = str(remainder) + result

      number //= base

    return result


def from_base(number, base):
    """!!!

    Результат:
        int: число в 10-й с.с."""
    number = number.upper()
    result = 0
    for i, digit in enumerate(reversed(number)):
      if digit in DIGITS_EX:
        result += DIGITS_EX[digit] * (base ** i)
      else:
        result += int(digit) * (base **i)

    return result


num_in_base10 = 255
base = 16

print(to_base(num_in_base10, base))

num_in_base16 = 'A'
print(from_base(num_in_base16, base))

# --------------
# Пример вывода:
#
# Нет
```

```python
# РПОС_ПР№4.2.11
# Выполнил: Городнянская М.С.
# Группа: МС-32



def ceasar(text, shift):
    """Вернуть измененную строку 'text' со сдвигом 'shift'.

    Параметры:
        - text (str): строка;
        - shift (int): свдиг.

    Результат:
        str: измененная строка."""
    # Набор кириллических букв
    letters = [chr(i) for i in range(ord('а'), ord('я') + 1)]
    lettersBig = [chr(i) for i in range(ord('А'), ord('Я') + 1)]

    res = ''
    for char in text:
      if char in letters:
        index = letters.index(char)
        res += letters[(index + shift) % len(letters)]
      elif char in lettersBig:
        index = lettersBig.index(char)
        res += lettersBig[(index + shift) % len(lettersBig)]
      else:
        res += char

    return res


text = input('Введите предложение: ')
shift = int(input('Введите сдвиг: '))

encoded = ceasar(text, shift)
decoded = ceasar(encoded, -shift)
print("Зашифрованная строка:", encoded)
print("Расшифрованная строка:", decoded)

# --------------
# Пример вывода:
#
# Введите предложение: ПрограММиРОВание С++
# Введите сдвиг: 4
# Зашифрованная строка: УфтзфдРРмФТЖдсмй Х++
# Расшифрованная строка: ПрограММиРОВание С++
```

```python
# РПОС_ПР№4.2.12
# Выполнил: Городнянская М.С.
# Группа: МС-32



# В данной задаче ввод с клавиатуры не нужен.
#
# Используйте пример данных ниже, при необходимости измените его для
# проверки правильности решения

data = [
    {1: 'м', 2: 'м', 3: 'м', 4: 'ж'},
    {1: 'ж', 2: 'м', 3: 'ж', 4: 'ж'},
    {1: 'ж', 2: 'ж', 3: 'ж', 4: 'ж'},
    {1: 'м', 2: 'м', 3: 'м', 4: 'м'},
    {1: None, 2: None, 3: None, 4: None},
    {1: 'м', 2: None, 3: None, 4: 'ж'},
    {1: None, 2: None, 3: None, 4: None},
    {1: 'м', 2: 'м', 3: None, 4: 'м'},
    {1: 'ж', 2: None, 3: None, 4: 'ж'}
]


def vacant_compartments(data):
    """Вернуть список полностью свободных купе. Нумерация купе идет с 1.

    Параметры:
        - data (list of dict): данные о занятости мест в вагоне.

    Результат:
        list of int."""
    res = []
    for compartment_number, compartment in enumerate(data, start = 1):
      empty_seat = 0
      for seat in compartment:
        if compartment[seat] is None:
          empty_seat += 1
      if empty_seat == 4:
        res.append(compartment_number)

    return res


def vacant_seats(data, compartments_condition=None, seat_condition=None):
    """Вернуть список свободных мест при соблюдении условий
    'compartments_condition' и 'seat_condition'.
    Нумерация купе и мест идет с 1.

    Параметры:
        - data (list of dict): данные о занятости мест в вагоне;
        - compartments_condition (function):
          функция c 1 параметром (словарь - купе), возвращающая True/False;
        - seat_condition (function):
          функция c 2 параметрами (номер места, значение),
          возвращающая True/False.

    Результат:
        list of tuple: список кортежей вида (номер купе, номер места)."""
    res = []
    for compartment_number, compartment in enumerate(data, start = 1):
      for seat_number, status in compartment.items():
        if status is None:
          if (compartments_condition is None or
              compartments_condition(compartment)):
              if seat_condition is None or seat_condition(seat_number, status):
                res.append((compartment_number, seat_number))

    return res


def is_same_sex_and_vacant(compartment, sex):
    """Вернуть True, если в купе 'compartment' есть свободные места,
    а остальные пассажиры только пола 'sex'.

    Параметры:
        - compartment (dict): данные о купе;
        - sex (str): пол ("м" или "ж").

    Результат:
        bool."""
    count_male = 0
    count_female = 0
    count_None = 0
    for seat_number, status in compartment.items():
      if status == "м":
        count_male += 1
      if status == "ж":
        count_female += 1
      if status is None:
        count_None += 1

    if count_None > 0:
      if sex == "м":
        return (count_male > 0 and count_male < 4 and count_female == 0)
      if sex == "ж":
        return (count_female > 0 and count_female < 4 and count_male == 0)


# список полностью свободных купе
print(vacant_compartments(data))
# список свободных мест в вагоне
print(vacant_seats(data))
# список свободных нижних мест
print(vacant_seats(data, seat_condition=lambda seat, value: seat % 2 != 0))
# список свободных верхних мест
print(vacant_seats(data, seat_condition=lambda seat, value: seat % 2 == 0))
# список свободных мест в купе с исключительно мужской компанией
print(vacant_seats(data, lambda x: is_same_sex_and_vacant(x, "м")))
# список свободных мест в купе с исключительно женской компанией
print(vacant_seats(data, lambda x: is_same_sex_and_vacant(x, "ж")))


# --------------
# Пример вывода:
#
# [5, 7]
# [(5, 1), (5, 2), (5, 3), (5, 4), (6, 2), (6, 3), (7, 1), (7, 2), (7, 3),
#  (7, 4), (8, 3), (9, 2), (9, 3)]
# [(5, 1), (5, 3), (6, 3), (7, 1), (7, 3), (8, 3), (9, 3)]
# [(5, 2), (5, 4), (6, 2), (7, 2), (7, 4), (9, 2)]
# [(8, 3)]
# [(9, 2), (9, 3)]
```

```python
# РПОС_ПР№4.2.13
# Выполнил: Городнянская М.С.
# Группа: МС-32



# В данной задаче ввод с клавиатуры не нужен.
#
# Используйте значения 'votes', при необходимости измените его для
# проверки правильности решения

votes = [
    2, 3, -1, 2, 5, 1, 1, 4, 1, 1, 1, 3, 1, 3, 5, -1, 5, 2, 5, 5,
    3, 3, 2, 3, 3, 2, 2, 1, 3, 2, 5, 2, 2, 4, 1, 1, 3, 2, 2, 3, 3,
    3, 1, 4, 2, 1, 4, 2, 3, 3, 3, -1, 5, 3, 1, 4, 5, 1, 1, 3, 3,
    3, -1, 5, 3, 3, 2, 5, 1, 1, 5, -1, 1, 2, 2, 3, -1, 4, 2, 5, 4,
    2, -1, 3, 1, 4, 3, 5, 4, 1, 5, 3, 2, 4, 2, 1, 3, 4, 1, 1, 3, 4
]

parties = {
    1: "Партия №1", 2: "Партия №2", 3: "Партия №3",
    4: "Партия №4", 5: "Партия №5", -1: "Испорчено"
}


def parties_votes(parties, votes):
    """Вернуть информацию о голосах 'votes', отданных за партии 'parties'.
    Испорченные бланки также присутствуют в результате.

    Параметры:
        - parties (dict): информация о партиях (номер голоса: название);
        - votes (list): номера голосов.

    Результат:
        dict: название: кол-во отданных голосов."""
    res = {}
    count_vote = 0
    for num_vote, name in parties.items():
      for vote in votes:
        if num_vote == vote:
          count_vote += 1
      res[name] = count_vote
      count_vote = 0

    return res


def print_results(votes_for_p):
    """Вывести результаты голосования в формате:

    1. Партия №2 | 1111 | 58.21%
    2. Партия №4 |  999 | 38.14%

    Параметры:
        - votes_for_p (dict): результат функции parties_votes().
    """
    sum_votes = 0
    for _, count_vote in votes_for_p.items():
      sum_votes += count_vote

    sorted_dict = sorted(votes_for_p.items(), key=lambda item: item[1], reverse=True)

    for i, (name, count_votes) in enumerate(sorted_dict):
      print(f'{i+1}. {name} | {count_votes:2} |',
            f'{round((count_votes * 100 / sum_votes), 2):5}%')


print_results(parties_votes(parties, votes))

# --------------
# Пример вывода:
#
# 1. Партия №3 | 27 | 26.47%
# 2. Партия №1 | 22 | 21.57%
# 3. Партия №2 | 20 | 19.61%
# 4. Партия №5 | 14 | 13.73%
# 5. Партия №4 | 12 | 11.76%
# 6. Испорчено |  7 |  6.86%
```
